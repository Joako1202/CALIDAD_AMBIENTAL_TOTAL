📑 Parte 1: Lectura y Validación de Estándares (ECAs)
🎯 Objetivo:
Cargar el archivo maestro ECAs_data.xlsx que contiene los parámetros de calidad ambiental permitidos para AIRE, SUELO, AGUA, etc., y asegurar que todos los valores sean numéricos.

🔍 Objetivo:
Carga masiva: Se lee el Excel completo (sheet_name=None) para obtener todas las pestañas.

Limpieza de datos: Se recorre cada hoja y se fuerzan las columnas de límites a ser números, eliminando textos o símbolos como ** que suelen venir en normativas.

Identificación: Se separan los nombres de los parámetros de los valores de los límites.


```python
import pandas as pd

# 1. Nombre de tu archivo Excel tal cual lo tienes en tu carpeta
archivo_excel = "ECAs_data.xlsx" 

try:
    # 2. Leer todas las hojas del Excel a la vez
    # sheet_name=None carga todas las pestañas en un diccionario
    biblioteca_hojas = pd.read_excel(archivo_excel, sheet_name=None)
    
    print(f"--- EXCEL LEÍDO: {archivo_excel} ---")
    print(f"Hojas detectadas: {list(biblioteca_hojas.keys())}\n")

    # 3. Procesar cada hoja por separado para validar números
    for nombre_hoja, df in biblioteca_hojas.items():
        # Identificamos columnas que no son texto
        cols_omitir = ['Parametros', 'Periodo', 'Zona', 'Unidad']
        cols_datos = [c for c in df.columns if c not in cols_omitir]
        
        # Validación Numérica: Limpiar '**', espacios o textos
        for col in cols_datos:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        
        print(f"✅ Hoja '{nombre_hoja}' lista.")
        print(f"   Columnas de límites: {cols_datos}")

except FileNotFoundError:
    print(f"❌ Error: No se encontró el archivo '{archivo_excel}' en tu carpeta.")
    print("Asegúrate de que el nombre sea exacto y que el archivo no esté abierto.")
```

Limpieza Automática: Convierte los valores a números, eliminando cualquier símbolo o texto extraño que traiga el Excel original.

```python
import pandas as pd

# 1. Cargar el Excel
archivo_excel = "ECAs_data.xlsx"
biblioteca_hojas = pd.read_excel(archivo_excel, sheet_name=None)

# 2. Selección de Hoja
print("--- SELECTOR DE ESTÁNDAR AMBIENTAL ---")
hojas_disponibles = list(biblioteca_hojas.keys())
for i, hoja in enumerate(hojas_disponibles):
    print(f"{i}: {hoja}")

eleccion_hoja = int(input("\nEscribe el NÚMERO de la hoja: "))
hoja_seleccionada = hojas_disponibles[eleccion_hoja]
df_crudo = biblioteca_hojas[hoja_seleccionada]

# Detectar identificador (Parametros o Zona)
columna_id = df_crudo.columns[0]

# 3. Selección Múltiple de Columnas
cols_omitir = [columna_id, 'Periodo', 'Zona', 'Unidad']
divisiones_disponibles = [c for c in df_crudo.columns if c not in cols_omitir]

print(f"\nDivisiones en {hoja_seleccionada}:")
for i, div in enumerate(divisiones_disponibles):
    print(f"{i}: {div}")

entrada_usuario = input("\nEscribe los NÚMEROS de las columnas (ej: 0,1): ")
indices = [int(x.strip()) for x in entrada_usuario.split(",")]
columnas_elegidas = [divisiones_disponibles[i] for i in indices]

# --- 4. CREACIÓN DEL DATAFRAME DE TRABAJO (df_final) ---
# Seleccionamos solo las columnas deseadas
df_final = df_crudo[[columna_id] + columnas_elegidas].copy()

# Convertimos las columnas de límites a números (limpieza de '**', 'N.A.', etc.)
for col in columnas_elegidas:
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')

print(f"\n✅ Dataframe 'df_final' creado para {hoja_seleccionada}.")
print(f"Dimensiones: {df_final.shape} (Filas, Columnas)")
print("-" * 30)

# Ahora puedes trabajar solo con df_final
print(df_final.head())
```

📂 Parte 4: Procesamiento Automatizado de CSV

Esta función estandariza archivos de datos con diferentes formatos en una tabla única y limpia.

✨ Capacidades principales:
Lectura Inteligente: Detecta automáticamente el separador (coma o punto y coma) y soporta caracteres especiales (tildes/ñ).

Mapeo Automático: Identifica columnas mediante palabras clave (ej: "Estación", "Lugar", "Fecha", "Conc").

Limpieza de Datos: * Convierte textos a números (corrige comas decimales).

Normaliza formatos de fecha.

Salida Estándar: Devuelve siempre un DataFrame con 4 columnas: Estaciones, Fecha, Parametro y Valores.


```python
import pandas as pd

def procesar_csv_inteligente(nombre_archivo):
    print(f"--- ANALIZANDO: {nombre_archivo} ---")
    try:
        # 1. Cargar archivo
        df = pd.read_csv(nombre_archivo, sep=None, engine='python', encoding='latin-1')
        df.columns = df.columns.str.strip()

        # 2. Buscador de Estaciones/Puntos
        col_est = next((c for c in df.columns if any(k in c.lower() for k in ['estac', 'punto', 'cod', 'lugar'])), None)
        if col_est:
            df = df.rename(columns={col_est: 'Estaciones'})
        else:
            df['Estaciones'] = 'No especificado'

        # 3. Buscador de Valores (Numérico)
        col_val = next((c for c in df.columns if any(k in c.lower() for k in ['valor', 'result', 'conc'])), 'Valores')
        df[col_val] = pd.to_numeric(df[col_val].astype(str).str.replace(',', '.'), errors='coerce')
        df = df.rename(columns={col_val: 'Valores'})

        # 4. TU LÓGICA: Buscador de Fecha y Parámetro
        col_fecha = next((c for c in df.columns if 'fecha' in c.lower()), None)
        col_para = next((c for c in df.columns if any(k in c.lower() for k in ['parametro', 'parámetro', 'variable'])), None)

        if col_fecha and col_para:
            df[col_fecha] = pd.to_datetime(df[col_fecha], dayfirst=True, errors='coerce')
            df = df.rename(columns={col_fecha: 'Fecha', col_para: 'Parametro'})
            print(f"✅ Mapeo exitoso: Estaciones, Valores, 'Fecha' y 'Parametro'")
        else:
            print("⚠️ Revisa los nombres de las columnas: No se encontró Fecha o Parámetro.")

        return df[['Estaciones', 'Fecha', 'Parametro', 'Valores']]
    
    except Exception as e:
        print(f"❌ Error al procesar {nombre_archivo}: {e}")
        return None

print("⚙️ Función configurada para todos los CSV.")
```

📈 Parte 5: Cruce de Datos y Análisis de Excedencias

🔍 Etapas del Análisis:
Preparación Dinámica: Identifica automáticamente todas las columnas de límites disponibles en la tabla de estándares para compararlas contra los valores medidos.

Cruce de Información (Merge): Une las tablas de Muestreo y Límites utilizando el nombre del parámetro como punto de unión.

Cálculo de Estadísticos: Para cada parámetro, el sistema identifica:

Valores máximos y mínimos (con su fecha y estación respectiva).

Total de muestras analizadas.

Detección de Excedencias: Calcula cuántas veces se superó cada límite específico, generando el conteo de excedencias y el porcentaje (%) de incumplimiento de forma dinámica.

📋 Salidas de Análisis:
Genera una tabla final (reporte_final) donde cada fila representa un parámetro y cada columna muestra su comportamiento estadístico frente a los diferentes límites configurados.


```python
# 1. Preparar las columnas de límites dinámicamente
col_id_eca = df_final.columns[0]
# Tomamos todas las columnas excepto la primera (que es el nombre del parámetro)
cols_limites_elegidas = [c for c in df_final.columns if c != col_id_eca]

# 2. El Cruce (Merge) - Ahora incluimos todas las columnas elegidas
df_unificado = pd.merge(
    df_muestreo, 
    df_final, # Aquí ya van el ID y todos los límites seleccionados
    left_on='Parametro', 
    right_on=col_id_eca, 
    how='inner'
)

# 3. Función de Estadísticos Adaptada para Múltiples Límites
def generar_reporte_multi(grupo):
    idx_max = grupo["Valores"].idxmax()
    idx_min = grupo["Valores"].idxmin()
    
    # Creamos un diccionario base con los datos generales
    resultado = {
        "Máximo": grupo.loc[idx_max, "Valores"],
        "Fecha Máx": grupo.loc[idx_max, "Fecha"].date() if pd.notnull(grupo.loc[idx_max, "Fecha"]) else "S/F",
        "Estación Máx": grupo.loc[idx_max, "Estaciones"],
        "Mínimo": grupo.loc[idx_min, "Valores"],
        "Total Muestras": len(grupo)
    }

    # Ahora calculamos excedencias para CADA columna de límite seleccionada
    for col in cols_limites_elegidas:
        val_limite = grupo[col].iloc[0]
        # Contamos cuántos valores superan este límite específico
        n_exc = (grupo["Valores"] > val_limite).sum() if pd.notna(val_limite) else 0
        
        # Agregamos los resultados al diccionario con el nombre del límite
        resultado[f"Límite {col}"] = val_limite
        resultado[f"Exc. {col}"] = n_exc
        resultado[f"% Exc. {col}"] = round((n_exc / len(grupo)) * 100, 2)

    return pd.Series(resultado)

# 4. Ejecución del reporte
# Usamos include_groups=False para evitar advertencias de versiones futuras de Pandas
reporte_final = df_unificado.groupby("Parametro", group_keys=False).apply(generar_reporte_multi, include_groups=False)

print(f"✅ Reporte generado comparando contra: {cols_limites_elegidas}")
display(reporte_final)
```

📂 Parte 6: Preparación del Entorno y Librerías

🎯 Objetivo:
Cargar las herramientas necesarias para la manipulación de datos y la generación de gráficos de alta calidad.

🔍 ¿Qué estamos usando?
Pandas: Para manejar las tablas de datos (DataFrames) de manera eficiente.

Matplotlib: Nuestra herramienta principal para dibujar los gráficos.

OS: Para interactuar con el sistema de archivos (crear carpetas y rutas).

```python
import os
import matplotlib.pyplot as plt
import pandas as pd


def ejecutar_reporte_grafico_final(df, carpeta_destino="Graficos_Finales"):
    ruta_completa = os.path.join(os.getcwd(), carpeta_destino)
    if not os.path.exists(ruta_completa):
        os.makedirs(ruta_completa)
    
    t_titulo, t_eje_y, t_leyenda = 16, 12, 9
    parametros = df['Parametro'].unique()

    for p in parametros:
        df_plot = df[df['Parametro'] == p].copy().sort_values('Fecha')
        if df_plot.empty: continue

        n_datos = len(df_plot)
        # Ajuste dinámico de etiquetas
        if n_datos > 50: t_fuente_x, rotacion = 6, 90
        elif n_datos > 20: t_fuente_x, rotacion = 8, 45
        else: t_fuente_x, rotacion = 10, 0
            
        ancho_dinamico = max(12, n_datos * 0.4) 

        # Formato de fecha
        meses_es = {'Jan':'Ene', 'Feb':'Feb', 'Mar':'Mar', 'Apr':'Abr', 'May':'May', 'Jun':'Jun', 
                    'Jul':'Jul', 'Aug':'Ago', 'Sep':'Sep', 'Oct':'Oct', 'Nov':'Nov', 'Dec':'Dic'}
        def fmt(dt):
            if pd.isnull(dt): return "S/F"
            f = dt.strftime('%b-%y')
            for eng, esp in meses_es.items(): f = f.replace(eng, esp)
            return f

        df_plot['Etiqueta_X'] = df_plot.apply(lambda x: f"{x['Estaciones']}\n({fmt(x['Fecha'])})", axis=1)

        fig, ax = plt.subplots(figsize=(ancho_dinamico, 7))
        
        # --- CAMBIO: Ancho de barra 0.8 para mejor estética ---
        ax.bar(df_plot['Etiqueta_X'], df_plot['Valores'], color='#EAEAEA', edgecolor='#4D4D4D', label='Resultado', width=0.8)

        # Líneas de Límites
        cols_meta = ['Estaciones', 'Fecha', 'Parametro', 'Valores', 'Parametros', 'Unidad', 'index', 'Etiqueta_X']
        limites_detectados = [c for c in df.columns if c not in cols_meta]
        estilos = [('#D62728', '-.'), ('#FF7F0E', '--'), ('#2CA02C', ':')]
        
        for i, col_limite in enumerate(limites_detectados):
            val_limite = df_plot[col_limite].iloc[0]
            if pd.notna(val_limite):
                color, estilo = estilos[i % len(estilos)]
                ax.axhline(y=val_limite, color=color, linestyle=estilo, linewidth=2, label=f"{col_limite}: {val_limite}")

        # --- MEJORA CLAVE: ELIMINAR ESPACIO EN BLANCO A LOS LADOS ---
        ax.set_xlim(-0.6, n_datos - 0.4) 
        
        # Formatos
        ax.set_title(f"Concentración de {p}", fontsize=t_titulo, fontweight='bold', pad=20)
        ax.set_ylabel("Concentración (mg/kg)", fontsize=t_eje_y)
        plt.xticks(rotation=rotacion, ha='right' if rotacion != 0 else 'center', fontsize=t_fuente_x)
        ax.legend(loc='upper right', fontsize=t_leyenda, frameon=True, edgecolor='gray')
        ax.grid(axis='y', linestyle=':', alpha=0.5)
        ax.set_axisbelow(True)

        plt.tight_layout()

        nombre_seguro = "".join([c for c in str(p) if c.isalnum() or c in (' ', '-', '_')]).strip()
        path_final = os.path.join(ruta_completa, f"Grafico_{nombre_seguro}.png")
        
        plt.savefig(path_final, dpi=300, bbox_inches='tight')
        plt.close()
        print(f"✅ Guardado sin espacios: {path_final}")

ejecutar_reporte_grafico_final(df_unificado)
```
